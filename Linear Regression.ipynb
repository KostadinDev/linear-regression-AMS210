```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [2,4,6,8,10,12]
accidents = [10,8,3,8,4,5]
x = np.array(years)
y = np.array(accidents)
ones = np.ones(x.size)
```

```python
q =  x.dot(y)/x.dot(x)
y_predicted = x.dot(q)
y_predicted
```

```python
plt.plot(x,y, 'o-',label = 'true')
plt.plot(x,y_predicted, 'o-',label = 'predicted')
plt.xlabel('Years')
plt.ylabel("Accidents")
plt.legend()
```

```python
SSD = np.sum((y_predicted-y)**2)
SSD
```

# y_predicted = qx + r

```python
q = (len(x)*x.dot(y) - x.dot(ones)*y.dot(ones))/(len(x)*(x**2).dot(ones) - (x.dot(ones))**2)
r = (y.dot(ones)*(x**2).dot(ones)-x.dot(ones)*(x.dot(y)))/ (len(x)*(x**2).dot(ones) - x.dot(ones)**2)
print(f"q = {q}, r = {r}]")
```

```python
y_predicted = q*x+r
y_predicted
```

```python
plt.plot(x,y, 'o-',label = 'true')
plt.plot(x,y_predicted, 'o-',label = 'predicted')
plt.xlabel('Years')
plt.ylabel("Accidents")
plt.legend()
```

```python
((y_predicted-y)**2).dot(ones)
```
