```python
# libraries for plotting and matrix operations
import matplotlib.pyplot as plt
import numpy as np
```

# Section 4.2, Problem 2

```python
# Data from problem
years = [2,4,6,8,10,12]
accidents = [10,8,3,8,4,5]
x = np.array(years)
y = np.array(accidents)
ones = np.ones(x.size)
```

```python
# part a)
q =  x.dot(y)/x.dot(x)
y_predicted = x.dot(q)
print(f"y = {q}x")
```

```python
# Plot the line
plt.plot(x,y, 'o',label = 'true')
plt.plot(x,y_predicted, '-',label = 'predicted')
plt.xlabel('Years')
plt.ylabel("Accidents")
plt.title("Part a)")
plt.legend()
```

```python
# Sum of squared error
SSD = np.sum((y_predicted-y)**2)
print(f"Sum of squared error = {SSD}")
```

```python
# part b)
q = (len(x)*x.dot(y) - x.dot(ones)*y.dot(ones))/(len(x)*(x**2).dot(ones) - (x.dot(ones))**2)
r = (y.dot(ones)*(x**2).dot(ones)-x.dot(ones)*(x.dot(y)))/ (len(x)*(x**2).dot(ones) - x.dot(ones)**2)
y_predicted = q*x+r
print(f"q = {q}, r = {r}")
print(f"y = {q:.2f}x + {r:.2f}")
```

```python
plt.plot(x,y, 'o',label = 'true')
plt.plot(x,y_predicted, '-',label = 'predicted')
plt.xlabel('Years')
plt.ylabel("Accidents")
plt.title("Part b)")
plt.legend()
```

```python
# part c
x = x-x.mean()
q = (len(x)*x.dot(y) - x.dot(ones)*y.dot(ones))/(len(x)*(x**2).dot(ones) - (x.dot(ones))**2)
r = (y.dot(ones)*(x**2).dot(ones) - x.dot(ones)*x.dot(y))/(6*(x**2).dot(ones) - x.dot(ones)**2)
print(f"q = {q}, r = {r}")
print(f"y = {q:.2f}x + {r:.2f}")
```

```python
plt.plot(x,y, 'o',label = 'true')
plt.plot(x,y_predicted, '-',label = 'predicted')
plt.xlabel('x\'')
plt.ylabel("y")
plt.title("Part c)")
plt.legend()
```
